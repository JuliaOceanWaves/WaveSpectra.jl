```julia
using Unitful
using DimensionfulAngles
using DimensionfulAngles.DefaultSymbols
using Plots

push!(LOAD_PATH, "../src")
using WaveSpectra
```

```julia
Nf = 200
Δf = 0.1Hz
f = (0:Nf-1) * Δf

Nθ = 36
Δθ = 360° / Nθ
θ = (0:Nθ-1) * Δθ

S = Spectrum(randn(Nf, Nθ) * m^2 / Hz / °, f, θ)
```
