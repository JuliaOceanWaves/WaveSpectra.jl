```julia
using Plots
using CairoMakie

using Unitful
using DimensionfulAngles
using DimensionfulAngles.DefaultSymbols

push!(LOAD_PATH, "../src")
using WaveSpectra
```

```julia
Nf = 10
Δf = 0.1Hz
f = (Δf:Δf:Δf*Nf)

Nθ = 36
Δθ = 360° / Nθ
θ = (0:Nθ-1) * Δθ

S = Spectrum(abs.(randn(Nf, Nθ)) * m^2 / Hz / °, f, θ);
Sc = Spectrum(abs.(randn(Nf, Nf)) * m^2 / Hz^2, f, f);
So = integrate(S, :direction);
```

```julia
plot_spectrum(So)
```

```julia
f = Figure()
ax = Axis(f[1, 1])
plot_spectrum!(ax, So; ax_properties=Dict(:xlabel=>"AAA",), color=:red)
f
```

```julia
plot_spectrum(Sc)
```

```julia
plot_spectrum(S; colormap=:viridis)
```

```julia
f = Figure()
ax_p = PolarAxis(f[1, 1])
ax_c = Axis(f[1, 2])
plot_spectrum!(ax_p, S; ax_properties=Dict(:rticklabelsize=>8))
plot_spectrum!(ax_c, Sc; ax_properties=Dict(:aspect=>DataAspect()), colormap=:viridis)
f
```
